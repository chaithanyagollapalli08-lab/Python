Using Identifiers

* %d for integer
* %f for float
* %s for string

```python
f=20
print('may age is %d '%f)
```

```python
course='python'
time=6
duration=1.30
print('I am learning %s in viharatech for %d months with %d hours daily'%(course,time,duration))
```

using f and{}

```python
course='python'
time=6
duration=1.30
print(f'I am learning {course} in viharatech for {time}  months with {duration} hours daily')
```
