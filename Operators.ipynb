```python
x = 5
x =x+ 3
print("x + 3 →", x)

x = 10
x =x- 4
print("x - 4 →", x)

x = 4
x =x* 2
print("x * 2 →", x)

x = 9
x =x/ 3
print("x / 3 →", x)

x = 10
x =x//3
print("x // 3 →", x)

x = 10
x =x% 4
print("x % 4 →", x)

x = 2
x =x**3
print("x **3 →", x)
```

Bitwise Operators

```python
a = 5   # 0101
b = 3   # 0011

print("5 & 3 =", a & b) # if two bits are 1 return  1 otherwise 0
print("5 | 3 =", a | b) # if two bits are 0 return 0 otherwise 1
print("5 ^ 3 =", a ^ b) #if two bits are different return 1 otherwise o
print("~5 =", ~a) #~(5+1)
print("5 << 1 =", a << 1) # shift bits to left 1 time
print("5 >> 1 =", a >> 1) #shift bits to right 1 time
```
