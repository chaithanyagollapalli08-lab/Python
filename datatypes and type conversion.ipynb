```python
#Python treats values as True or False in a Boolean context:

#False values: 0, 0.0, "" (empty string), [], {}, set(), None

#Everything else is True
print(bool(0))
print(bool(0))      # False
print(bool(42))     # True
print(bool(""))     # False
print(bool("Hi"))   # True
print(bool([]))     # False
print(bool([1, 2, 3]))  # True
print(bool({}))     # False
```

**Type conversion**

int to float

```python
a=int(input())
print(a)
print(type(a))
b=float(a)
print(b)
print(type(b))
```

```python
a=int(input())
print(a)
print(type(a))
b=str(a)
print(b)
print(type(b))
```
