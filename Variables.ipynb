** Variables **
A variable is a name that stores a value in Python.

```python
a=10
print(a)
```

```python
a+10
```

**Rules for writing variable**
 * Variable name can include letters, digits, and underscores
  * Must start with a letter or underscore
  * Cannot start with a number
   * Cannot use spaces
  * Variable names are case-sensitive

Examples:

```python
age = 25
first_name = "John"
num1 = 100
_total = 200
print(age,first_name,num1,_total)
```

```python
1name = "John"   # starts with number
first name = "A" # space not allowed
class = 10       # class is a keyword
```

```python
a, b, c = 10, 20, 30
print(a,b,c)
```

```python
x = y = z = 100
print(x)
print(y)
print(z)
```

```python
#Changing Variable Values
x = 10
print(x)
x = 20  # updated value
print(x)
```
